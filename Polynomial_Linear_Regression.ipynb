## Polynomial Linear Regression

### Polynomia Linear Regression Genel Formülü


y = a + b1x + b2x^2 + b3x^3 + b4x^4 + .......... bN*x^4

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

df = pd.read_csv("polynomial.csv", sep=";")
```

```python
plt.scatter(df["deneyim"], df["maas"])
plt.xlabel("Deneyim(yıl)")
plt.ylabel("Maaş")
plt.savefig("1.png", dpi=300)
plt.show()
```

##### Bu model yanlış olcağı için polynomial kullanılcak

 a + b1x + b2x^2

```python
polynomial_regression = PolynomialFeatures(degree=4) #degree = N değeri
x_polynomial = polynomial_regression.fit_transform(df[["deneyim"]])
```

```python
reg = LinearRegression()
reg.fit(x_polynomial,df["maas"])
```

```python
y_head = reg.predict(x_polynomial)
plt.plot(df["deneyim"], y_head, color="red", label="polynomial regression")
plt.legend()

plt.scatter(df["deneyim"],df["maas"])
plt.show()
```

```python
x_polynomial1 = polynomial_regression.fit_transform([[4.5]])
maas_hesaplama = reg.predict(x_polynomial1)
```

```python
hesaplama = maas_hesaplama.copy()
```

```python
hesaplama = [round(x, 2) for x in hesaplama]
for hesap in hesaplama:
    print("{:,}".format(hesap))
```
